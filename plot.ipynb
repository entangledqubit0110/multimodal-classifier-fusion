```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
modalityNames = ["ECO", "SOC", "HMC", "NAC"]
```

```python
# heuristic scheme
train_acc_heu = 0.9
modality_wts_heu = [0.2762979, 0.3086317, 0.26821327, 0.14685714]
test_acc_heu = 1
```

```python
# random scheme
train_acc_rand = 0.7
modality_wts_rand = [0.13484396, 0.31784446, 0.34238777, 0.20492381]
test_acc_rand = 0.66
```

```python
# uniform scheme
train_acc_uni = 0.7
modality_wts_uni = [0.19301995, 0.01104898, 0.44519965, 0.35073143]
test_acc_uni = 0.75
```

```python
fig = plt.figure(figsize = (6, 4))
plt.bar(['heuristic', 'random', 'uniform'], [train_acc_heu, train_acc_rand, train_acc_uni], width=0.4, color='maroon')
plt.xlabel("feature importance scheme for labelling data points")
plt.ylabel("training accuracy")
plt.title("Training accuracy in different labelling policies")
plt.show()
```

```python
fig = plt.figure(figsize = (10, 6))
barWidth = 0.2
br1 = np.arange(len(modalityNames))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]


plt.bar(br1, modality_wts_heu, width=barWidth, color='r', label= 'heuristic')
plt.bar(br2, modality_wts_rand, width=barWidth, color='b', label= 'random')
plt.bar(br3, modality_wts_uni, width=barWidth, color='g', label= 'uniform')


plt.ylabel("weights of modality as given by model")
plt.xlabel("modality")
plt.xticks([r + barWidth for r in range(len(modalityNames))], modalityNames)
plt.title("Modality weights in different labelling policies")
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize = (6, 4))
plt.bar(['heuristic', 'random', 'uniform'], [test_acc_heu, test_acc_rand, test_acc_uni], width=0.4, color='crimson')
plt.xlabel("feature importance scheme for labelling data points")
plt.ylabel("test accuracy")
plt.title("Test accuracy in different labelling policies")
plt.show()
```
